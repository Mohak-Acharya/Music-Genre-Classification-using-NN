```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
DATA = pd.read_csv('Genre_data.csv')
DATA.head()
```

```python
DATA.shape
```

```python
DATA.dtypes
```

```python
DATA = DATA.drop(['filename','length'],axis=1)
```

```python
my_genre_list = DATA.iloc[:, -1]
le = LabelEncoder()
```

```python
y = le.fit_transform(my_genre_list)
```

```python
print(y)
```

```python
print(DATA.iloc[:, :-1])
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(DATA.iloc[:, :-1], dtype = float))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# Neural network
model = Sequential()
model.add(keras.layers.Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
classifier = model.fit(X_train,y_train,epochs=200,batch_size=64)
```

```python
test_loss, test_acc  = model.evaluate(X_test, y_test, batch_size=64)
```

## Accuracy on the training dataset is 0.9994 and on the test dataset is 0.9036
